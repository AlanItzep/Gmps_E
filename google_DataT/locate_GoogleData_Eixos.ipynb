```python
df_locate = table_original[['title','street','number']]
```

```python
two_chara = df_locate.street.apply(lambda x: x[:2])
```

```python
uniqs_two = two_chara.unique()
va_lst= []
for i in uniqs_two:
    if i.isupper() == True:
        if len(i) > 1:
            va_lst.append(i)
        
print(va_lst)
```

```python
df_locate['street'] = df_locate['street'].str.replace('CL ','Carrer ')
df_locate['street'] = df_locate['street'].str.replace('PZ ','Plaça ')
df_locate['street'] = df_locate['street'].str.replace('PS ','Passeig ')
df_locate['street'] = df_locate['street'].str.replace('AV ','Avinguda ')
df_locate['street'] = df_locate['street'].str.replace('RB ','Rambla ')
df_locate['street'] = df_locate['street'].str.replace('GV ','Gran via ')

df_locate['serch_o'] = df_locate['street']+', '+df_locate['number']
```
